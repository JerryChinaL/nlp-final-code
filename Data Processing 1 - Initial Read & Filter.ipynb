### Description of Notebook

This notebook reads the entire 5GB news data set and the price dataset from huggingface, and filters out tech stocks that our project focuses on. It then saves these filtered datasets as "tech_news.csv" and "tech_prices.csv" in the Data folder

```python
import pandas as pd
import numpy as np
```

```python
all_news = pd.read_csv("Data/News.csv")
```

```python
all_news['Date'] = pd.to_datetime(all_news['Date'], format='%Y-%m-%d %H:%M:%S %Z', utc=True)
```

```python
all_news["Date"][0]
```

```python
all_news.head()
```

```python
# List of tech stocks competing with Apple
tech_rivals = ['AAPL', 'ADBE', 'AMD', 'AMZN', 'BBK', 'GOOGL', 'GRMN', 'HPQ', 'IBM', 'INTC', 'NFLX', 'NVDA', 'ORCL', 'TSLA']

# Filter the DataFrame
tech_news = all_news[all_news["Stock_symbol"].isin(tech_rivals)]

tech_news.head()
```

```python
tech_news["Date"].min()
```

```python
tech_news.to_csv("Data/tech_news.csv")
```

```python
all_prices_df = pd.concat(
    [
        pd.read_csv(f"Data/Prices/{ticker}.csv").assign(
            Stock_symbol=ticker,
            Date=lambda df: pd.to_datetime(df["date"], errors="coerce")  # Auto-detect format
        )
        for ticker in tech_rivals
    ],
    ignore_index=True
)

filtered_prices_df = all_prices_df[all_prices_df["Date"] >= '08/15/2009']

filtered_prices_df
```

```python
filtered_prices_df["Date"].min()
```

```python
filtered_prices_df.to_csv("Data/tech_prices.csv")
```
